```python
# Load data
# data = np.load('/content/Y.npy')
import h5py
import numpy as np

file_path = 'datasets/BJ13_M32x32_T30_InOut.h5'

with h5py.File(file_path, 'r') as f:
    data = np.array(f['data'])
    date = np.array(f['date'])


data = data[:, 1, :, :]
print(data.shape)
```

```python
pred_len = 6
seq_len = 128
T = seq_len + pred_len
n_test = int(len(data) * 0.15)
# train_data = data[:-n_test]
test_data = data[-n_test:]
all_data =  []
for i in range(len(test_data)):
    if i + T < len(test_data):
        all_data.append(test_data[i:i+T].tolist())
save = {
    'X_test': [all_data]
}

import json
with open(f'TaxiBJout_{seq_len}_{pred_len}_short.json', 'w') as f:
    json.dump(save, f)
```
