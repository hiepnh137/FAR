```python
# Nearest neibhbor interpolation for missing values
import torch
import torch.nn.functional as F

x = torch.randn(32, 32)  # your tensor

x_up = F.interpolate(x.unsqueeze(0).unsqueeze(0), 
                     size=(256, 256), 
                     mode='nearest').squeeze()

print(x_up.shape)  # torch.Size([256, 256])

```

```python
# visualize x and x_up
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.title("Original x")
plt.imshow(x, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Upsampled x_up")
plt.imshow(x_up, cmap='gray')

plt.show()
```
