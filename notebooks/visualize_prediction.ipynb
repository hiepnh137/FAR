```python
import pickle as pkl
import os
import json
import numpy as np

seq_len = 16
pred_len = 6

path = '../results/TaxiBJout_16_6_FAR_pred.pkl'
data = pkl.load(open(path, 'rb'))

path = '../datasets/TaxiBJout_16_6_short.json'
raw = json.load(open(path, 'r'))

d = np.array(raw['X_test'][0])

history = d[:,:seq_len,:,:]

prediction = np.concatenate((history, data), axis=1)
print('prediction: ', prediction.shape)
```

```python
sample_id = 1
pix_x = 19
pix_y = 18

# visualize prediction and raw at sample_id, pix_x, pix_y
import matplotlib.pyplot as plt
plt.plot(range(seq_len + pred_len), prediction[sample_id, :, pix_x, pix_y], label='prediction')
plt.plot(range(seq_len + pred_len), d[sample_id, :, pix_x, pix_y], label='raw')
plt.axvline(x=seq_len-1, color='r', linestyle='--', label='prediction start')
plt.xlabel('Time step')
plt.ylabel('Value')
plt.title(f'Pixel ({pix_x}, {pix_y}) Prediction vs Raw')
plt.legend()
plt.show()
```
